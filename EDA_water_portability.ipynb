```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load data
df = pd.read_csv('data/water_potability.csv')
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
df.columns = df.columns.str.strip()

# -----------------------------------------
# 1. BASIC DATA OVERVIEW
# -----------------------------------------
print("=" * 60)
print("DATASET OVERVIEW")
print("=" * 60)
print(f"Shape: {df.shape}")
print(f"\nColumn Names:\n{df.columns.tolist()}")
print(f"\nData Types:\n{df.dtypes}")
print(f"\nFirst 5 rows:\n{df.head()}")

# -----------------------------------------
# 2. MISSING VALUES ANALYSIS
# -----------------------------------------
print("\n" + "=" * 60)
print("MISSING VALUES ANALYSIS")
print("=" * 60)
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing,
    'Percentage': missing_pct
})
print(missing_df[missing_df['Missing_Count'] > 0].sort_values('Percentage', ascending=False))

# Visualize missing values
plt.figure(figsize=(10, 5))
missing_df[missing_df['Missing_Count'] > 0].plot(kind='bar', y='Percentage')
plt.title('Missing Values Percentage')
plt.ylabel('Percentage (%)')
plt.xlabel('Features')
plt.tight_layout()
plt.show()

# -----------------------------------------
# 3. BASIC STATISTICS
# -----------------------------------------
print("\n" + "=" * 60)
print("DESCRIPTIVE STATISTICS")
print("=" * 60)
print(df.describe())

# -----------------------------------------
# 4. CLASS DISTRIBUTION (POTABILITY)
# -----------------------------------------
print("\n" + "=" * 60)
print("CLASS DISTRIBUTION (POTABILITY)")
print("=" * 60)
class_dist = df['Potability'].value_counts()
class_pct = (df['Potability'].value_counts(normalize=True) * 100)
print(f"Not Potable (0): {class_dist[0]} ({class_pct[0]:.2f}%)")
print(f"Potable (1): {class_dist[1]} ({class_pct[1]:.2f}%)")
print(f"Imbalance Ratio: {class_dist[0]/class_dist[1]:.2f}:1")

# Plot class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Potability', data=df, palette='Set1')
plt.title('Water Potability Distribution')
plt.xlabel('Potability')
plt.ylabel('Count')
plt.show()

# -----------------------------------------
# 5. DISTRIBUTION PLOTS (HISTOGRAMS & KDE)
# -----------------------------------------
print("\n" + "=" * 60)
print("FEATURE DISTRIBUTIONS")
print("=" * 60)

features = df.drop('Potability', axis=1).columns
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for idx, feature in enumerate(features):
    axes[idx].hist(df[feature].dropna(), bins=30, edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'Distribution of {feature}')
    axes[idx].set_xlabel(feature)
    axes[idx].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# KDE plots
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for idx, feature in enumerate(features):
    sns.kdeplot(data=df, x=feature, ax=axes[idx], shade=True)
    axes[idx].set_title(f'KDE of {feature}')

plt.tight_layout()
plt.show()

# -----------------------------------------
# 6. CORRELATION MATRIX
# -----------------------------------------
print("\n" + "=" * 60)
print("CORRELATION ANALYSIS")
print("=" * 60)
corr_matrix = df.corr()
print(corr_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

# Top correlations with Potability
potability_corr = corr_matrix['Potability'].sort_values(ascending=False)
print(f"\nFeatures correlated with Potability:\n{potability_corr}")

# -----------------------------------------
# 7. BOXPLOTS BY POTABILITY
# -----------------------------------------
print("\n" + "=" * 60)
print("BOXPLOTS BY POTABILITY")
print("=" * 60)

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for idx, feature in enumerate(features):
    sns.boxplot(x='Potability', y=feature, data=df, ax=axes[idx], palette='Set2')
    axes[idx].set_title(f'{feature} by Potability')

plt.tight_layout()
plt.show()

# -----------------------------------------
# 8. OUTLIER DETECTION (IQR METHOD)
# -----------------------------------------
print("\n" + "=" * 60)
print("OUTLIER DETECTION (IQR METHOD)")
print("=" * 60)

def detect_outliers_iqr(data, feature):
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[feature] < lower) | (data[feature] > upper)]
    return len(outliers), lower, upper

outlier_summary = []
for feature in features:
    count, lower, upper = detect_outliers_iqr(df, feature)
    outlier_summary.append({
        'Feature': feature,
        'Outlier_Count': count,
        'Lower_Bound': lower,
        'Upper_Bound': upper
    })
    print(f"{feature}: {count} outliers (bounds: {lower:.2f} - {upper:.2f})")

# -----------------------------------------
# 9. PAIRPLOT (FEATURE INTERACTIONS)
# -----------------------------------------
print("\n" + "=" * 60)
print("PAIRPLOT (FEATURE INTERACTIONS)")
print("=" * 60)

# Sample data for pairplot (if too large)
sample_df = df.sample(min(1000, len(df)), random_state=42)
sns.pairplot(sample_df, hue='Potability', diag_kind='kde', palette='Set1')
plt.suptitle('Pairplot of Features', y=1.00)
plt.tight_layout()
plt.show()

# -----------------------------------------
# 10. SKEWNESS & KURTOSIS
# -----------------------------------------
print("\n" + "=" * 60)
print("SKEWNESS & KURTOSIS")
print("=" * 60)

skew_kurt = pd.DataFrame({
    'Feature': features,
    'Skewness': [df[f].skew() for f in features],
    'Kurtosis': [df[f].kurtosis() for f in features]
})
print(skew_kurt)

print("\nEDA Complete! Summary saved.")
```
